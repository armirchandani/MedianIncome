Importing dependencies

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

Data Processing


```python
# loading the csv data to a Pandas DataFrame
df = pd.read_csv('/content/emp_08.csv')
```

```python
df
```

```python
miami_dade_df = df[df["geo_name"] == "Miami-Dade County, Florida"]
```

```python
print(miami_dade_df )
```

```python
# Data from Miami-Dade County
races = [
    "Overall",
    "Black or African American",
    "White",
    "Hispanic or Latino",
    "Asian"
]
income_values = [
    68694,   # Overall median income
    51550,   # Black alone
    77378,   # White alone
    67018,   # Hispanic
    92788    # Asian alone
]
```

```python
# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(races, income_values, color=['gray', 'black', 'blue', 'orange', 'green'])
plt.title("Median Income by Race/Ethnicity - Miami-Dade County (2023)")
plt.ylabel("Median Income (USD)")
plt.xticks(rotation=45)

# Add value labels on top
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 1000, f"${yval:,}", ha='center', va='bottom')

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
